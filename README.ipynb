**Preparing the Data (single-folder layout)**

Put the dataset files next to your notebooks/code (same directory), for example:

train_cleaned.csv

test_cleaned.csv

Put the best_trial files next to your notebooks/code (same directory), for example:

best_model_trial_0.pt

best_model_trial_3.pt

Put the checkpoints folders next to your notebooks/code (same directory), for example:
